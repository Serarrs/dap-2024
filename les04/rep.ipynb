Задача №5. Выведите распределение переменной Pclass (социально-экономический статус) и это же распределение, только для мужчин / женщин по отдельности. Сколько было мужчин 2-го класса

```python
import numpy as np
import pandas as pd
url= 'https://raw.githubusercontent.com/chebotarevsa/dap-2024/refs/heads/main/les04/data/titanic.csv'
df = pd.read_csv(url)
print("Общее распределение Pclass:")
print(df['Pclass'].value_counts(), "\n")
print("Распределение Pclass по полу:")
print(df.groupby('Sex')['Pclass'].value_counts(), "\n")
men_2nd_class = df[(df['Sex'] == 'male') & (df['Pclass'] == 2)].shape[0]
print(f"Количество мужчин 2-го класса: {men_2nd_class}")
```

Задача №6. Каковы медиана и среднеквадрати́чное отклонение платежей (Fare)? Округлите до 2 десятичных знаков.

```python
import numpy as np
import pandas as pd
url= 'https://raw.githubusercontent.com/chebotarevsa/dap-2024/refs/heads/main/les04/data/titanic.csv'
df = pd.read_csv(url)
median_fare = round(df['Fare'].median(), 2)
std_fare = round(df['Fare'].std(), 2)
print("медиана и среднеквадрати́чное отклонение платежей:")
median_fare, std_fare
```

Задача №7. Правда ли, что люди моложе 30 лет выживали чаще, чем люди старше 60 лет? Каковы доли выживших в обеих группах?

```python
import numpy as np
import pandas as pd
url= 'https://raw.githubusercontent.com/chebotarevsa/dap-2024/refs/heads/main/les04/data/titanic.csv'
df = pd.read_csv(url)
young = df[df['Age'] < 30]
old = df[df['Age'] > 60]

young_surv_rate = round(young['Survived'].mean(), 2)
old_surv_rate = round(old['Survived'].mean(), 2)
print("Выживших среди молодых и старых:")
young_surv_rate, old_surv_rate
```

Задача №9. Найдите самое популярное имя среди пассажиров Титаника мужского пола?

```python
import numpy as np
import pandas as pd
url= 'https://raw.githubusercontent.com/chebotarevsa/dap-2024/refs/heads/main/les04/data/titanic.csv'
df = pd.read_csv(url)
men = df[df['Sex'] == 'male']
import re
from collections import Counter
def extract_name(fullname):
    match = re.search(r'Mr\. ([^()]*)', fullname)
    if match:
        name_part = match.group(1).strip()
        first_name = name_part.split()[0].replace('"', '').replace("'", '')
        return first_name
    return None

men_names = men['Name'].apply(extract_name).dropna()
Counter(men_names).most_common(1)
```
