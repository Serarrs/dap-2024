1. Типы ячеек в Jupyter Notebook.
<br>
Markdown (текстовы) ячейки - используется для оформления текста
<br>
Python (кодовые) ячейки - используется для написания и выполнения кода
2. Как с помощью Markdown сформировать иерархический список?
<br>
Иерархический (вложенный) список создаётся с помощью отступов, а также маркеров
3. Как с помощью Markdown оформить цитату?
<br>
Цитата создаётся с помощью символа ">"
4. Как в текстовую ячейку вставить отформатированный блок кода?
<br>
Используйте тройные обратные кавычки (```) до и после блока кода
5. Сколько уровней заголовка существует в Markdown?
<br>
6 уровенй заголовков (#, ##, ###, ####, #####, ######)
6. Шаги процесса анализа данных, рассказать подробно про один из шагов.
<br>
Определение проблемы или выдвижение идей, Извлечение или сбор данных, Подготовка данных, Исследование и визуализация данных, Предсказательная модель, Проверка модели, Развертывание - финальный шаг процесса анализа, задача которого — предоставить результаты, то есть выводы анализа. В процессе развертывания бизнес-среды анализ является выгодой, которую получит клиент, заказавший анализ.
7. Типы данных, рассказать подробно про один из типов данных.
<br>
Структурированные данные - это тип данных, при котором последние упорядочены в вертикальные столбцы (поля) и горизонтальные строки (записи или наблюдения), Неструктурированные данны, Полуструктурированные данные
8. Что такое Jupyter Notebook, для чего используется?
<br>
Jupyter Notebook - интерактивный блокнот, первоначально являвшийся веб-реализацией и развитием IPython, ставший самостоятельным проектом, ориентированным на работу со множеством сред выполнения.
9. Почему python популярен для анализа данных?
<br>
Простоте синтаксиса, Мощным библиотекам (pandas, numpy, matplotlib, scikit-learn), Большому сообществу, Возможности интеграции с другими инструментами (SQL, R, Spark и др.)
10. Обязательно ли использовать Jupyter Notebook для анализа данных?
Нет, не обязательно. Хотя Jupyter удобен для интерактивного анализа, также можно использовать: IDE (например, PyCharm, VS Code), Терминал/консоль Python, Скрипты .py, Другие среды (например, RStudio, Databricks)
